```python
#https://scikit-learn.org/stable/
import pandas as pd
import glob
from sklearn.ensemble import RandomForestClassifier
```

```python
all_files = glob.glob('data/*.csv')
```

```python
df = pd.concat((pd.read_csv(f) for f in all_files)).reset_index()
#for each file in all_files, read csv and concatenate into dataframe
```

```python
df_indicators = df[[
    'covid19_test_results',
    'high_risk_exposure_occupation',
    'cough',
    'sob',
    'diarrhea',
    'fatigue',
    'loss_of_smell',
    'loss_of_taste',
    'runny_nose',
    'muscle_sore',
    'sore_throat'
]]
```

```python
#pd.options.mode.chained_assignment = None
```

```python
df_indicators.dropna(inplace=True)
```

```python
df_indicators.describe()
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(df_indicators.iloc[:,1:], df_indicators['covid19_test_results'])
```

```python
df_indicators.iloc[:,1:]
```

```python
feature_df = pd.DataFrame({'feature': df_indicators.columns[1:],
                          'importance': clf.feature_importances_})
```

```python
feature_df.sort_values(by='importance', ascending=False)
```
